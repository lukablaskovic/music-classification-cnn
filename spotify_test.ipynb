# Spotify data loader

```python
from tensorflow.keras.models import load_model
import os
import numpy as np
from tensorflow.keras.preprocessing.image import img_to_array, load_img

genres = ['blues', 'classical', 'country', 'disco', 'hiphop', 'jazz', 'metal', 'pop', 'reggae', 'rock']
genre_to_label = {genre: index for index, genre in enumerate(genres)}

def load_and_preprocess_images(base_folder, genres, grayscale=True, target_size=(256, 256)):
    images = []
    labels = []
    filenames = []

    print(f"Loading images from {base_folder}...")
    for genre in genres:
        genre_folder = os.path.join(base_folder, genre)
        if not os.path.isdir(genre_folder):
            print(f"Warning: {genre_folder} is not a directory.")
            continue

        for filename in os.listdir(genre_folder):
            image_path = os.path.join(genre_folder, filename)
            if os.path.isfile(image_path):
                if grayscale:
                    image = load_img(image_path, target_size=target_size, color_mode='grayscale')
                    print(f"Loaded {filename} from {genre} in grayscale mode.")
                else:
                    image = load_img(image_path, target_size=target_size)
                    print(f"Loaded {filename} from {genre} in color mode.")
                image = img_to_array(image)
                image /= 255.0  # Normalizing to [0, 1]
                images.append(image)
                labels.append(genre_to_label[genre])  # Using genre_to_label for consistent labeling
                filenames.append(filename)
            else:
                print(f"Skipped {filename} as it is not a file.")

    print(f"Finished loading {len(images)} images.")
    return np.array(images), np.array(labels), filenames


folder_path = 'dataset/spotify_mp3_resampled_grayscale'

X_new, y_new, filenames_new = load_and_preprocess_images(folder_path, genres, grayscale=True)

print("X_new.shape", X_new.shape)
print("y_new.shape", y_new.shape)
```

```python
from tensorflow.keras.models import load_model
import os
import numpy as np
from tensorflow.keras.preprocessing.image import img_to_array, load_img

model_path = "final_final/model24.h5"
model = load_model(model_path)

predictions = model.predict(X_new)
predicted_labels = np.argmax(predictions, axis=1)

loss, accuracy, top_k = model.evaluate(X_new, y_new, verbose=1)

print(f"Test Loss: {loss}")
print(f"Test Accuracy: {accuracy}")
print(f"Test Top-k Accuracy: {top_k}")
print("______________________________")
for filename, predicted_label, actual_label in zip(filenames_new, predicted_labels, y_new):
    predicted_genre = genres[predicted_label]
    actual_genre = genres[actual_label]
    print(f"{filename}: Predicted - {predicted_genre}, Actual - {actual_genre}")
```
